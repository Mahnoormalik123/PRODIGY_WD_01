## Bar plot for Categorical variable: region wise population distribution

```python
# Select the dataset of year 2023
latest_population_data = df[['Country Name', 'Country Code', '2023']]

# merge the data of region wise population
region_population_data = df.merge(metadata, how='left', left_on='Country Code', right_on='Country Code')

# calculate the sum of region wise population
region_population_sum = region_population_data.groupby('Region')['2023'].sum().reset_index()

# Bar plot for region-wise population distribution
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='2023', data=region_population_sum)
plt.xlabel('Region')
plt.ylabel('Population in 2023')
plt.title('Region-wise Population Distribution in 2023')
plt.xticks(rotation=45)
plt.show()
```

### Region-wise Population Distribution in 2023

This bar chart shows the population of different regions in the year 2023.

- **East Asia & Pacific**: Has the highest population, around 2 billion people.
- **South Asia**: The second highest, with about 1.8 billion people.
- **Europe & Central Asia** and **Latin America & Caribbean**: These regions have moderate populations, around 0.7 billion and 0.6 billion respectively.
- **Middle East & North Africa**: Has a population of about 0.4 billion.
- **North America**: The lowest population, around 0.3 billion.
- **Sub-Saharan Africa**: Around 1 billion people, less than East Asia & Pacific and South Asia, but more than other regions.

## Histogram for continuous variable:

### Population distribution in Pakistan

```python
# filter the data of Pakistan
pakistan_data = df[df['Country Name'] == 'Pakistan']

# prepare the data of year wise population
years = [str(year) for year in range(1960, 2024)]
pakistan_population = pakistan_data[years].melt(var_name='Year', value_name='Population')

# Histogram plot for Pakistan population distribution over years
plt.figure(figsize=(12, 6))
sns.histplot(pakistan_population['Population'], bins=30, kde=True)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution in Pakistan (1960-2023)')
plt.show()
```

### Population Distribution in Pakistan (1960-2023)

This histogram shows how Pakistan's population has changed from 1960 to 2023.

- **Frequency Representation**: Each bar shows how many years the population was in a certain range.
- **Population Growth Trend**: The first few bars are higher because the population was lower in the early years (1960s). As time goes on, the population increases, but the years with these higher populations are fewer.
- **Distribution Shape**: The bars are mostly on the left, showing that the population has been growing over time.

## Histogram for Distribution of Population in 2023:

```python
# Selecting population data for the year 2023
population_2023 = df['2023']

# Plot the histogram
plt.figure(figsize=(12, 6))
plt.hist(population_2023.dropna(), bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Population in 2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Distribution of Population in 2023

This histogram shows the distribution of populations across different countries in the year 2023.

- **Population**: The total population of each country in 2023.
- **Frequency**: Indicates how many countries fall into each population category.

The histogram helps identify the most common population sizes and the range of population sizes across countries in 2023.

#### Analysis

1. **High Frequency of Low Populations**:
   - The first bar represents countries with populations close to zero (but greater than zero).
   - The frequency is very high, indicating that most countries have relatively small populations.

2. **Decreasing Frequency with Higher Populations**:
   - As the population increases, the frequency of countries within those population ranges decreases significantly.
   - There are very few countries with populations over 1 billion.

3. **Outliers**:
   - There are a few bars representing populations up to 8 billion, but these are extremely rare.
   - The outliers may represent highly populous countries like China and India.
